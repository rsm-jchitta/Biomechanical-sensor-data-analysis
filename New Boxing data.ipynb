```python
import numpy as np

def create_segments_with_test_id_overlap(df, x=21):
    # List to store segments and their labels
    segments = []
    labels = []
    
    # Calculate step size for 50% overlap
    step_size = x // 2  # Integer division to ensure an integer result

    # Iterate through each unique class label and test identifier
    for motion in df['Motion'].unique():
        for test_id in df[df['Motion'] == motion]['Test_num'].unique():
            test_df = df[(df['Motion'] == motion) & (df['Test_num'] == test_id)]
            
            # Iterate to create segments of 'x' rows with the calculated step size for overlap
            for start in range(0, len(test_df) - x + 1, step_size):  # Ensure the loop goes to the last possible segment
                end = start + x
                segment = test_df.iloc[start:end]
                
                # No need to check if segment has 'x' rows since the loop condition ensures it
                segment_data = segment.drop(['Motion', 'time', 'Test_num'], axis=1).values
                segments.append(segment_data)
                labels.append(motion)

    # Convert lists to numpy arrays
    segments_array = np.array(segments)
    labels_array = np.array(labels)

    return segments_array, labels_array

# Example usage
x_value = 85  # Segment size
segments, labels = create_segments_with_test_id_overlap(df, x=x_value)

# 'segments' now contains the feature segments with 50% overlap,
# and 'labels' contains the corresponding class labels.
```

```python
import numpy as np
import scipy.fftpack

def extract_features(segments):
    features_list = []

    for segment in segments:
        # Statistical features
        means = np.mean(segment, axis=0)
        std_devs = np.std(segment, axis=0)
        variances = np.var(segment, axis=0)
        mins = np.min(segment, axis=0)
        maxs = np.max(segment, axis=0)
        medians = np.median(segment, axis=0)
        
        # Frequency-domain features using FFT
        fft_coeffs = scipy.fftpack.fft(segment, axis=0)
        fft_coeffs_real = np.real(fft_coeffs)
        fft_coeffs_imag = np.imag(fft_coeffs)
        
        # For simplicity, let's compute the mean of the absolute values of the FFT coefficients
        fft_abs_mean = np.mean(np.abs(fft_coeffs), axis=0)
        
        # Combine all features into a single array
        features = np.concatenate([means, std_devs, variances, mins, maxs, medians, fft_abs_mean])
        features_list.append(features)
    
    return np.array(features_list)

# Assuming your segments are stored in a variable named `segments`
features = extract_features(segments)

print("Shape of first feature vector:", features[0].shape)
```

```python
def generate_feature_names(num_features):
    feature_names = []
    stats = ['mean', 'std_dev', 'variance', 'min', 'max', 'median']
    
    for i in range(num_features):
        for stat in stats:
            feature_names.append(f"Feature_{i+1}_{stat}")
    
    # Assuming the FFT absolute mean is calculated across all features together
    feature_names.extend([f"Feature_{i+1}_fft_abs_mean" for i in range(num_features)])
    
    return feature_names

# Example: For a dataset with 2 features (e.g., 2 sensors)
num_features = 3
feature_names = generate_feature_names(num_features)
print(feature_names)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Example: Training a Random Forest Classifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
# Define the parameter grid
param_grid = {
    'n_estimators': [100],
    'max_depth': [10]
}

# Initialize the classifier
rf = RandomForestClassifier(random_state=42)

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, scoring='accuracy', verbose=1)

# Fit GridSearchCV
grid_search.fit(X_train, y_train)

# Best parameters and best score
print("Best parameters:", grid_search.best_params_)
print("Best score:", grid_search.best_score_)

# Use the best estimator on the validation set
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
val_accuracy = accuracy_score(y_test, y_pred)
print(f"Test Accuracy: {val_accuracy}")


# Generate classification report
print(classification_report(y_test, y_pred))

```
